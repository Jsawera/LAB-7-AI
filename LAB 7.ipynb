```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = pd.read_csv("income.csv")    # Load the dataset

data.columns = data.columns.str.strip()    # Clean column names (if needed)

print("Dataset:")  # Display the first few rows of the dataset
print(data.head())

k = 3   # Define the number of clusters

kmeans = KMeans(n_clusters=k, random_state=0)   # Perform K-Means Clustering
data['Cluster'] = kmeans.fit_predict(data[['Age', 'Income']])  # Use both Age and Income for clustering

centroids = kmeans.cluster_centers_   # Display the centroids
print("\nCentroids:")
for i, centroid in enumerate(centroids, start=1):
    print(f"Centroid {i}: Age = {centroid[0]:.2f}, Income = {centroid[1]:.2f}")

plt.figure(figsize=(8, 6))     # Visualize the clusters
for cluster in range(k):
    cluster_data = data[data['Cluster'] == cluster]
    plt.scatter(cluster_data['Age'], cluster_data['Income'], label=f'Cluster {cluster + 1}')

for i, centroid in enumerate(centroids):  # Plot the centroids
    plt.scatter(centroid[0], centroid[1], s=200, c='black', marker='x', label=f'Centroid {i + 1}')

plt.xlabel('Age')
plt.ylabel('Income')
plt.title('K-Means Clustering')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = {    # Create the dataset
    'Objects': ['OB-1', 'OB-2', 'OB-3', 'OB-4', 'OB-5', 'OB-6', 'OB-7', 'OB-8'],
    'X': [1, 1, 1, 2, 1, 2, 1, 2],
    'Y': [4, 2, 4, 1, 1, 4, 2, 1],
    'Z': [1, 2, 2, 2, 1, 2, 1, 1]
}

df = pd.DataFrame(data) # Convert the dataset into a pandas DataFrame

coordinates = df[['X', 'Y', 'Z']]   # Drop the 'Objects' column for clustering purposes

# Perform K-Means Clustering with K=2
kmeans = KMeans(n_clusters=2, random_state=0)
df['Cluster'] = kmeans.fit_predict(coordinates)

# Display the centroids
centroids = kmeans.cluster_centers_
print("Centroids of the clusters:")
for i, centroid in enumerate(centroids, start=1):
    print(f"Centroid {i}: X = {centroid[0]:.2f}, Y = {centroid[1]:.2f}, Z = {centroid[2]:.2f}")

# Visualize the clusters
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for each cluster
for cluster in range(2):
    cluster_data = df[df['Cluster'] == cluster]
    ax.scatter(cluster_data['X'], cluster_data['Y'], cluster_data['Z'], label=f'Cluster {cluster + 1}')

# Plot the centroids
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2], s=200, c='black', marker='x', label='Centroids')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D K-Means Clustering')
ax.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules

# Task 1: Segmenting Patients Based on Age and BMI using K-Means Clustering
data = pd.DataFrame({   # Load the patient data (Age and BMI)
    'Patient ID': ['P-1', 'P-2', 'P-3', 'P-4', 'P-5', 'P-6', 'P-7', 'P-8'],
    'Age': [25, 40, 30, 60, 50, 28, 70, 55],
    'BMI': [22.5, 28.0, 26.5, 30.2, 32.1, 23.8, 31.4, 29.0]
})
# Perform K-Means clustering to segment patients into two clusters (K=2)
k = 2
kmeans = KMeans(n_clusters=k, random_state=0)
data['Cluster'] = kmeans.fit_predict(data[['Age', 'BMI']])

centroids = kmeans.cluster_centers_  # Display the cluster centroids
print("\nCluster Centroids:")
for i, centroid in enumerate(centroids, start=1):
    print(f"Centroid {i}: Age = {centroid[0]:.2f}, BMI = {centroid[1]:.2f}")

# Visualize the clusters
plt.figure(figsize=(8, 6))
for cluster in range(k):
    cluster_data = data[data['Cluster'] == cluster]
    plt.scatter(cluster_data['Age'], cluster_data['BMI'], label=f'Cluster {cluster + 1}')

# Plot the centroids
for i, centroid in enumerate(centroids):
    plt.scatter(centroid[0], centroid[1], s=200, c='black', marker='x', label=f'Centroid {i + 1}')

plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('K-Means Clustering: Patient Segmentation')
plt.legend()
plt.grid(True)
plt.show()

# Task 2: Identifying Frequent Medical Conditions Using the Apriori Algorithm

# Load the patient medical condition data (1 for presence, 0 for absence)
transactions = pd.DataFrame({
    'Diabetes': [1, 1, 0, 1, 1, 0, 1, 0],
    'Hypertension': [0, 1, 1, 1, 1, 1, 0, 1],
    'Heart Disease': [1, 0, 1, 0, 1, 0, 1, 0],
    'Asthma': [0, 0, 0, 1, 0, 1, 0, 1]
})

# Apply the Apriori algorithm with a minimum support of 5% (0.05)
frequent_itemsets = apriori(transactions, min_support=0.05, use_colnames=True)

# Display the frequent itemsets
print("\nFrequent Itemsets:")
print(frequent_itemsets)

# Generate the association rules from the frequent itemsets
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Display the generated association rules
print("\nAssociation Rules:")
print(rules)
```
